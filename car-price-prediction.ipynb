```python
cars.replace({'Fuel_Type':{'Petrol':0,'Diesel':1,'CNG':2}},inplace=True)
cars.replace({'Selling_type':{'Dealer':0,'Individual':1}},inplace=True)
cars.replace({'Transmission':{'Manual':0,'Automatic':1}},inplace=True)
```

```python
cars['Fuel_Type'].value_counts()
```

```python
X = cars.drop('Selling_Price', axis=1)
y = cars['Selling_Price']
```

```python
X.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=69)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
lin_reg = LinearRegression()
```

```python
lin_reg.fit(X_train, y_train)
```

```python
y_train_pred = lin_reg.predict(X_train)
y_test_pred = lin_reg.predict(X_test)
```

```python
print("Coefficient of Determination R2 on train set:", lin_reg.score(X_train, y_train))
print("Coefficient of Determination R2 on test set:", lin_reg.score(X_test, y_test))
```

```python
mse_train = mean_squared_error(y_train_pred, y_train)
mse_test = mean_squared_error(y_test_pred, y_test)

print("Mean Squared Error on train set:", mse_train)
print("Mean Squared Error on test set:", mse_test)
```

```python
mae_train = mean_absolute_error(y_train_pred, y_train)
mae_test = mean_absolute_error(y_test_pred, y_test)

print("Mean Absolute Error on train set:", mae_train)
print("Mean Absolute Error on test set:", mae_test)
```

```python
def train_eval_models(X_train, X_test, y_train, y_test, estimator):
    estimator.fit(X_train, y_train)
    
    y_train_pred = estimator.predict(X_train)
    y_test_pred = estimator.predict(X_test)
    
    print("Coefficient of Determination R2 on train set:", estimator.score(X_train, y_train))
    print("Coefficient of Determination R2 on test set:", estimator.score(X_test, y_test))
    
    print()
    print()
    
    mse_train = mean_squared_error(y_train_pred, y_train)
    mse_test = mean_squared_error(y_test_pred, y_test)

    print("Mean Squared Error on train set:", mse_train)
    print("Mean Squared Error on test set:", mse_test)
    
    print()
    print()
    mae_train = mean_absolute_error(y_train_pred, y_train)
    mae_test = mean_absolute_error(y_test_pred, y_test)

    print("Mean Absolute Error on train set:", mae_train)
    print("Mean Absolute Error on test set:", mae_test)
```

```python
from sklearn.linear_model import SGDRegressor, Ridge, Lasso
```

```python
model1 = SGDRegressor()
model2 = Ridge()
model3 = Lasso()
```

```python
print("Performance of Stochastic Gradient Descent Regressor: ")
train_eval_models(X_train, X_test, y_train, y_test, model1)
```

```python
print("Performance of Ridge Regressor: ")
train_eval_models(X_train, X_test, y_train, y_test, model2)
```

```python
print("Performance of Lasso Regressor: ")
train_eval_models(X_train, X_test, y_train, y_test, model3)
```
